# Data Cleaning :
1. Column Year has many non year values.
2. Data type of Year column should be int.
3. Many entries in the price column were either invalid, or unrealistic
4. The price column was found in object (string) format instead of numerical (int/float).
5. The price column contained unwanted characters such as commas (,) and symbols like $.
6. The kms_driven column was in string/object format instead of numeric (int).
7. The values contained the text "kms" along with the number.
8. Numbers had commas (,) in between, making them invalid for numerical conversion and has NaN values.
9. The fuel_type column has NaN values.
10. Keeping First three words of name column.

# MODEL

```python
X= df.drop(columns = "Price")
y = df["Price"]
```

```python
X 
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y ,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline 
```

```python
ohe = OneHotEncoder() # creating object of OneHotEncoder
ohe.fit(X[["name","company","fuel_type"]])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer(
    (OneHotEncoder(categories = ohe.categories_),["name","company","fuel_type"]),
    remainder = "passthrough"
)
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test , y_pred)
```

```python
score = []
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2 , random_state = i)
    le = LinearRegression
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    score.append(r2_score(y_test,y_pred))
```

```python
np.argmax(score)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2 , random_state = np.argmax(score))
le = LinearRegression
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
